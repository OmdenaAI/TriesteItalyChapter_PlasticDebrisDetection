```python
# add your GitHub and DagsHub username and authentication token
# and the email you want to use for git commits as secrets to the notebook
# you can do that from the right panel

from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
dagshub_username = user_secrets.get_secret("DAGSHUB_USERNAME")
dagshub_token = user_secrets.get_secret("DAGSHUB_TOKEN")
github_username = user_secrets.get_secret("GITHUB_USERNAME")
github_token = user_secrets.get_secret("GITHUB_TOKEN")
email = user_secrets.get_secret("EMAIL")
```
